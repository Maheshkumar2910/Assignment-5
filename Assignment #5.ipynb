**Function to Plot the line graph for the four datasets**

```python
def visualize(filename):
    "Plot the line graph"
    data = pd.read_csv(filepath_or_buffer = filename)
    Average = data.mean(axis=0)
    Maximum = data.max(axis=0)
    Minimum = data.min(axis=0)
    fig, (a1,a2,a3) = plt.subplots(1, 3, figsize =(25,7))
    a1.plot(Average,color='orange')
    a2.plot(Maximum, color = 'green')
    a3.plot(Minimum, color='Red')

    a1.set_xlabel("Feature#")
    a1.set_ylabel("Average")
    a1.set_title("Dataset")

    a2.set_xlabel("Feature#")
    a2.set_ylabel("Max")
    a2.set_title("Dataset")

    a3.set_xlabel("Feature#")
    a3.set_ylabel("Min")
    a3.set_title("Dataset")

    
    plt.show()
```

**Function to identify the issue of the 4 datasets**

```python
def Identify_issues(filename):
    "To identify the issue of the dataset"
    data = pd.read_csv(filepath_or_buffer = filename)
    Maximum = data.max(axis=0)
    Minimum = data.min(axis=0)
    if Maximum[0] == 0 and Maximum[20]==20:
        print("I doubt the integrity of this dataset")
    elif sum(Minimum) == 0:
        print("Why is this happening")
    else:
        print("Dataset looks good")
    
    
```

**Calling function for printing line graph and issue of datasets**

```python
#Calling functions to print the line graph and its corresponding issue of the dataset. 
for datasets in dataset[:4]:
    print("\n")
    print(datasets)
    visualize(datasets)
    Identify_issues(datasets)
```
